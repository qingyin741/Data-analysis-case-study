# 背景描述


在这个日益注重健康与体能的时代，健身已成为许多人追求健康生活的重要组成部分。  
本数据集包含若干健身房会员的详细信息，包括年龄、性别、体重、身高、心率、锻炼类型、身体脂肪比例等多项关键指标，旨在深入探索会员的健身习惯、生理状态及其与健康成果之间的联系。

# 数据说明  


| 字段 | 说明 |  
| --- | --- |  
| Age | 健身房会员的年龄。 |  
| Gender | 健身房会员的性别（男性或女性）。 |  
| Weight (kg) | 会员的体重（Kg）。 |  
| Height (m) | 会员的身高（m）。 |  
| Max_BPM | 锻炼期间的最大心率（每分钟心跳次数）。 |  
| Avg_BPM | 锻炼期间的平均心率。 |  
| Resting_BPM | 锻炼前的静息心率。 |  
| Session_Duration (hours) | 每次锻炼的持续时间（小时）。 |  
| Calories_Burned | 每次锻炼消耗的总卡路里。 |  
| Workout_Type | 进行的锻炼类型（例如，Cardio/有氧运动, Strength/力量训练, Yoga/瑜伽, HIIT/高强度间歇训练）。 |  
| Fat_Percentage | 会员的身体脂肪百分比。 |  
| Water_Intake (liters) | 锻炼期间的每日饮水量（升）。 |  
| Workout_Frequency (days/week) | 每周的锻炼次数。 |  
| Experience_Level | 经验水平，从 1 (初学者) 到 3 (专家)。 |  
| BMI | BMI指数，由身高和体重计算得出。 |

## 不同锻炼类型的平均卡路里消耗

```python
# 分组计算每种锻炼类型的平均卡路里消耗
calories_mean = df.groupby("Workout_Type")["Calories_Burned"].mean().sort_values()

# 绘制柱状图
colors = ["skyblue", "lightgreen", "salmon", "gold"]
ax = calories_mean.plot(kind="bar", color=colors, figsize=(8, 5), edgecolor='black')

# 中文标题和轴标签
plt.title("不同锻炼类型的平均卡路里消耗", fontsize=14)
plt.xlabel("锻炼类型", fontsize=12)
plt.ylabel("平均卡路里消耗", fontsize=12)
plt.xticks(rotation=0, fontsize=11)
plt.yticks(fontsize=11)

# 添加数值标签
for i, val in enumerate(calories_mean.values):
    plt.text(i, val + 5, f"{val:.0f}", ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


# 周锻炼时长 = 每次时长 * 每周频率
df["周锻炼时长"] = df["Session_Duration (hours)"] * df["Workout_Frequency (days/week)"]

# 卡路里强度指数 = 每小时卡路里消耗
df["卡路里强度指数"] = df["Calories_Burned"] / df["Session_Duration (hours)"]

# 心率活跃度 = Max_BPM / Resting_BPM
df["心率活跃度"] = df["Max_BPM"] / df["Resting_BPM"]

# 选择用于聚类的特征
features_for_clustering = [
    "Workout_Frequency (days/week)", "Session_Duration (hours)",
    "Calories_Burned", "Experience_Level", "周锻炼时长",
    "卡路里强度指数", "心率活跃度"
]

# 标准化数据
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[features_for_clustering])

# KMeans聚类
kmeans = KMeans(n_clusters=3, random_state=42)
df['聚类标签'] = kmeans.fit_predict(scaled_data)

# PCA降维用于可视化
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
df['PCA1'] = pca_data[:, 0]
df['PCA2'] = pca_data[:, 1]

# 可视化（Jupyter中查看）
plt.figure(figsize=(8, 6))
for label in df['聚类标签'].unique():
    subset = df[df['聚类标签'] == label]
    plt.scatter(subset['PCA1'], subset['PCA2'], label=f'群组 {label}')

plt.title("KMeans聚类结果 (PCA降维后)")
plt.xlabel("PCA 维度1")
plt.ylabel("PCA 维度2")
plt.legend()
plt.grid(True)
plt.show()

# 聚类标签人数统计
print(df["聚类标签"].value_counts())

# 每个群组的均值特征分析
cluster_means = df.groupby("聚类标签")[features_for_clustering].mean()
print("\n每个群组的均值特征：")
print(cluster_means)
```
